##Frauded Credit Card Detection
Main Purpose is to analyze if card transaction was frauded

At the very begin, we want to read data. Data is from Kaggle (https://www.kaggle.com/mlg-ulb/creditcardfraud). Using pandas lib:

```python
import pandas as pd
ccdata = pd.read_csv('creditcard.csv')
```

###Some data visualization

To avoid some problems with names in python libraries, I'm changing column 'Class' name to 'Output'

```python
ccdata.rename(columns={'Class': 'Output'}, inplace=True)
```

```python
ccdata.tail()
```

```python
ccdata.describe()
```

```python
from matplotlib import pyplot as plt
fig1, ax1 = plt.subplots()
ax1.set_title('Amount feature horizontal boxplot')
ax1.set_xlabel('Amount')
ax1.boxplot(ccdata.Amount, notch=True, vert=False)
plt.show()
```

Let's look at how output data (column 'Output') is divided

```python
fig2, ax2 = plt.subplots()
ax2.set_title('Output feature proportions')
ax2.set_xlabel('Output')
out_proportions = ccdata.groupby(['Output']).count()
ax2.bar(x=['Not frauded','Frauded'], height=out_proportions['Amount'])
plt.show()
```

As we can see, we have much less not frauded records:

```python
summary_out_proportions = out_proportions['Amount']/out_proportions['Amount'].sum()
print(summary_out_proportions)
```

Now, we will try to see if some of our input variables (Time, V1..V28, Amount) are related with themselves and with 'Output':

As you may see, we have also 'Time' column in our data, which is log of credit cards payments. We can easily see time series in this:

```python
fig3, ax3 = plt.subplots()
ax3.set_title('Time distribution')
ax3.set_xlabel('Time')
ax3.set_ylabel('Count')
ax3.hist(ccdata.Time, bins=100)
plt.show()
```

We can see diffrence of records count in time. It seems that frequency of transactions is divided into two subsets.
